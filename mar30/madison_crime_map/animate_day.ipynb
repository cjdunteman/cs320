```python
from matplotlib import pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.core.display import HTML, display
from time import time
import geopandas
```

```python
city = geopandas.read_file("zip://states.zip")
crime = geopandas.read_file("crime.shp")
crime
```

```python
fig, ax = plt.subplots(figsize=(6, 6))

ax.set_xlim(-89.5, -89.3)
ax.set_ylim(43, 43.2)

city.plot(color="0.85", ax=ax)

# List of (marker, percent)
markers = []
for idx, row in crime.iterrows():
    pt = row["geometry"]
    mkr = ax.plot(pt.x, pt.y, "ro")[0]
    markers.append((mkr, row["percent"]))

seconds = 12
fps = 2
frame_count = fps * seconds

def draw_frame(frame_num):
    percent = frame_num / frame_count
    for mkr, mkr_perc in markers:
        # if crime already happened at mkr
        if percent > mkr_perc:
            elapsed_time = (percent - mkr_perc) / 0.1
            alpha = max(1 - elapsed_time, 0)
        # if crime hasn't happended at mkr
        else:
            alpha = 0 # future
        mkr.set_alpha(alpha)
    
debug_frame = None

if debug_frame != None:
    draw_frame(debug_frame)
else:
    anim = FuncAnimation(fig, draw_frame, frames=frame_count, interval=1000/fps)
    t0 = time()
    html = anim.to_html5_video()
    t1 = time()
    print("Seconds to generate:", t1-t0)
    plt.close(fig)
    display(HTML(html))
```
