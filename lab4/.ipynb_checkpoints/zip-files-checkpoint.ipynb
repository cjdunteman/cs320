```python
from zipfile import ZipFile
```

```python
with ZipFile('mmt_gtfs.zip') as zf:
    for info in zf.infolist():
        print(info)
```

```python
# 

with ZipFile('mmt_gtfs.zip') as zf:
    for info in zf.infolist():
        orig_mb = info.file_size / (1024**2)
        ratio = info.file_size / info.compress_size
        s = "file {name:s}, {mb:.3f} MB (uncompressed), {ratio:.1f} compression ratio"
        print(s.format(name=info.filename, mb=orig_mb, ratio=ratio))
```

```python
# Overall compression ration

total_file_size = 0
total_compress_size = 0
with ZipFile('mmt_gtfs.zip') as zf:
    for info in zf.infolist():
        orig_mb = info.file_size / (1024**2)
        total_file_size += info.file_size 
        total_compress_size += info.compress_size
ratio = total_file_size / total_compress_size
ratio
```
